### 7.1 Global Minimum of Convex Function

Let $\mathbf{x}^*$ be the local minimum of $f$, assume it is not the global minimum, which meams there exist a $\mathbf{x}_0$ such that $f(\mathbf{x}_0) < f(\mathbf{x}^*)$.

By definition, for a convex function, for any $\theta \in (0,1)$ we have

$$
\begin{align*}
  f(\theta \mathbf{x}^* + (1-\theta)\mathbf{x}_0) &\le \theta f(\mathbf{x}^*) + (1-\theta)f(\mathbf{x}_0) \\
  &< \theta f(\mathbf{x}^*) + (1-\theta)f(\mathbf{x}^*)\\
  \Rightarrow f(\theta \mathbf{x}^* + (1-\theta)\mathbf{x}_0) &< f(\mathbf{x}^*)
\end{align*}
$$

Now, if $\theta\to 1$, $\theta \mathbf{x}^* + (1-\theta)\mathbf{x}_0 \to \mathbf{x}^*$, in such case, $\mathbf{x}^*$ cannot be the local minimum neither.

Therefore local minimun $\mathbf{x}^*$ of a convex function must also be the global minimum.
