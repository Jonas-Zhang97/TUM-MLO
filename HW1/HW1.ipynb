```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

create the required $x$ array

```python
x = np.arange(0, 1, 0.001)
print("Size of the x:", x.size)
```

genetate $\epsilon_i$ following gaussian distribution

```python
epsilon = np.random.normal(0, 0.25, 1000)
print("Size of the epsilon:", epsilon.size)
```

compute $y$

```python
y = x + epsilon
```

make the scatter plot

```python
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.savefig("/home/yueyangzhang/TUM/MLO/Problem4_1.png")
```

the minimum of $f(a) = \sum_{i=1}^{n} (x_i a -y_i)^2$ can be represented as $\text{argmin}_a \sum_{i=1}^{n} (x_i a -y_i)^2$, regarding to proposition 1 in lecture_note_1_2, a can be calculated as $(\underline{x}^\top \underline{x})^{-1} \underline{x}^\top \underline{y}$

```python
a = np.dot(pow(np.dot(x, x), -1), np.dot(x, y))
print("minimizer of f w.r.t. a is a =", a)
```

add the estimated linear model to the data

```python
g = np.dot(a, x)
plt.scatter(x, y)
plt.plot(x, g, color = "red")
plt.savefig("/home/yueyangzhang/TUM/MLO/Problem4_2.png")
```

as requested change the ground truth of $y$, denoted as $y_1$

```python
xi = np.random.normal(0, 0.01, 1000)

y_1 =  30 * ((x - 0.25) ** 2 * (x - 0.75) ** 2) + xi
plt.scatter(x, y_1)
```

For the new curve, applies the closed form of the least square solution

```python
a_1 = np.dot(pow(np.dot(x, x), -1), np.dot(x, y_1))
g_1 = a_1 * x
plt.scatter(x, y_1)
plt.plot(x, g_1, color = "red")
plt.savefig("/home/yueyangzhang/TUM/MLO/Problem4_4.png")
```
